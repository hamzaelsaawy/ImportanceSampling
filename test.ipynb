```julia
q1 = MixtureDistribution([Normal(2, 3), Cauchy(30), Laplace(-10)], [2, 3, 4])
q2 = MixtureDistribution([Poisson(1), Poisson(16)], [3, 10])
q3 = MixtureDistribution([Exponential(1), Exponential(2)], [3, 4])
;
```

```julia
q4 = MixtureDistribution([MvNormal([16.0, 16], eye(2)*3), MvTDist(6.0, [-16, -16.0], eye(2)*3)])
;
```

```julia
xs = rand(q1, 1000)
distplot(xs, kde=false, bins=150)
xlim(-50, 50)
;
```

```julia
xs = rand(q4, 1000)
jointplot(xs[1,:], xs[2,:])
;
```
